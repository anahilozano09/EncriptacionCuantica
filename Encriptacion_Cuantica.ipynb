```python
import random as r

bitsLista = []
basesAlice = []
qubitsAlice = []
basesBob = []
basesFinales = []
qubitsFinales = []
bitsFinales = []
bases = ['C', 'H']
basesEva = []
qubitsEva = []
qubitsBob = []
bits = 0

def menu():
    global bits
    print("Bienvenido a la simulación de Criptografía con Qubits")
    print("Nuestra simulación cuenta con dos opciones:")
    print("Para realizarlo sin metiche, marque 1. Si es con metiche, marque 2.")

    while True:
        opc = int(input("Ingrese su opción (1 o 2): "))
        if opc == 1:
            print("Has elegido la opción sin metiche.")
            bits = int(input("Ingrese la cantidad de bits con la que se va a trabajar: "))
            SinMetiche()
            break
        elif opc == 2:
            print("Has elegido la opción con metiche.")
            bits = int(input("Ingrese la cantidad de bits con la que se va a trabajar: "))
            Con_Metiche()
            break
        else:
            print("Ingrese solo valores válidos (1 ó 2)")

def regresarAlMenu():
    while True:
        print("¿Deseas regresar al menú?")
        print("Marca y para sí, n para no")
        salida = str(input())
        if salida == 'y':
            menu()
            break
        elif salida == 'n':
            print("Vuelve pronto!")
            break
        else:
            print("Ingrese solo valores válidos (y ó n)")

def generarQubits(bases, bitsLista):
    qubits = []
    for i in range(bits):
        if bitsLista[i] == 0:
            qubits.append('|0>' if bases[i] == 'C' else '|+>')
        else:
            qubits.append('|1>' if bases[i] == 'C' else '|->')
    return qubits

def SinMetiche():
    global bitsLista, basesAlice, qubitsAlice, basesFinales, qubitsFinales, bitsFinales
    basesIguales = True
    bitsLista = [r.randint(0, 1) for _ in range(bits)]
    print("Los bits generados son " + str(bitsLista))
    #Se generan las bases de Alice y Bob
    basesAlice = [r.choice(bases) for _ in range(bits)]
    basesBob = [r.choice(bases) for _ in range(bits)]
    print("La base Hadamard esta representada por H y la base computacional representada por C")
    print("Bases de Alice son " + str(basesAlice))
    #Se generan los qubits de Alice con ayuda de la función generarQubits
    qubitsAlice = generarQubits(basesAlice, bitsLista)
    print("Los qubits de Alice son " + str(qubitsAlice))
    print("Bases de Bob son " + str(basesBob))
    #Se buscan las bases que coinciden entre Alice y Bob, y se agregan a las listas finales
    for i in range(bits):
        if basesAlice[i] == basesBob[i]:
            basesFinales.append(basesAlice[i])
            qubitsFinales.append(qubitsAlice[i])
            bitsFinales.append(bitsLista[i])
        else:
            print("Las bases ubicadas en la posición " + str(i+1) + " no coinciden")
            print("La base de Alice es " + str(basesAlice[i]) + " y la de Bob es " + str(basesBob[i]))
            continue
    #Se verifican si existen bases que coincidan entre Alice y Bob
    if basesFinales == []:
        print("No se encontraron bases iguales entre Alice y Bob")
        basesIguales = False
    if basesIguales == True:
        print("Las bases y qubits obtenidos después de haber eliminado las que no coinciden son:")
        print("Las bases que coinciden con Alice y Bob son " + str(basesFinales))
        print("Los qubits que coinciden con Alice y Bob son " + str(qubitsFinales))
        print("La clave privada de Alice y Bob es " + str(bitsFinales))

    #Limpiamos las listas
    bitsLista.clear()
    basesAlice.clear()
    qubitsAlice.clear()
    basesBob.clear()
    basesFinales.clear()
    qubitsFinales.clear()
    bitsFinales.clear()
    regresarAlMenu()

def Con_Metiche():
    global bitsLista, basesAlice, qubitsAlice, basesEva, qubitsEva, basesBob, qubitsBob
    bitsLista = [r.randint(0, 1) for _ in range(bits)]
    basesIguales, mostrarClave = True, True
    print("Los bits generados son " + str(bitsLista))
    #Se generan las bases de Alice, Eva y Bob
    basesAlice = [r.choice(bases) for _ in range(bits)]
    basesBob = [r.choice(bases) for _ in range(bits)]
    basesEva = [r.choice(bases) for _ in range(bits)]
    print("La base Hadamard esta representada por H y la base computacional representada por C")
    print("Bases de Alice son " + str(basesAlice))
    #Se generan los qubits de Alice con ayuda de la función generarQubits
    qubitsAlice = generarQubits(basesAlice, bitsLista)
    print("Los qubits de Alice son " + str(qubitsAlice))

    print("Las bases de Eva son " + str(basesEva))
    qubitsEva = []
    #Se generan los qubits de Eva a partir de la comparación de las bases entre Alice y Eva
    for i in range(bits):
        probaQubits = r.random()
        if basesAlice[i] == basesEva[i]:
            qubitsEva.append(qubitsAlice[i])
        else:
            if basesEva[i] == 'C':
                qubitsEva.append('|0>' if probaQubits <= 0.5 else '|1>')
            elif basesEva[i] == 'H':
                qubitsEva.append('|+>' if probaQubits <= 0.5 else '|->')

    print("Los qubits de Eva son " + str(qubitsEva))
    print("Las bases de Bob son " + str(basesBob))
    qubitsBob, bitsListaCopia, basesAliceCopia, qubitsAliceCopia, basesEvaCopia, qubitsEvaCopia, basesBobCopia = [], [], [], [], [], [], []
    print("Las bases que no coinciden con Alice y Bob son:")
    #Se obtienen los bits, bases y qubits iguales entre Alice y Bob
    for i in range(bits):
        if (basesBob[i] == basesAlice[i]):
            bitsListaCopia.append(bitsLista[i])
            basesAliceCopia.append(basesAlice[i])
            qubitsAliceCopia.append(qubitsAlice[i])
            basesEvaCopia.append(basesEva[i])
            qubitsEvaCopia.append(qubitsEva[i])
            basesBobCopia.append(basesBob[i])
        else:
            print("Las bases ubicadas en la posición " + str(i+1))
            print("La base de Alice es " + str(basesAlice[i]) + " y la de Bob es " + str(basesBob[i]))

    #Si no se encuentran bases iguales
    if (bitsListaCopia == []):
        print("No se encontraron bases iguales entre Alice y Bob")
        basesIguales = False
    #Si se encuentran bases iguales
    if (basesIguales == True):
        print("Se encontraron bases iguales entre Alice y Bob")
        #Se imprimen las bases iguales entre Alice y Bob, al igual que las bases finales de Eva
        print("Las bases y qubits obtenidos después de haber eliminado las que no coinciden son:")
        print("Bases de Alice son " + str(basesAliceCopia))
        print("Los qubits de Alice son " + str(qubitsAliceCopia))
        print("Las bases de Eva son " + str(basesEvaCopia))
        print("Los qubits de Eva son " + str(qubitsEvaCopia))
        print("Las bases de Bob son " + str(basesBobCopia))
        #Se obtienen los qubits de Bob
        for i in range(len(bitsListaCopia)):
            probaQubits = r.random()
            if basesBobCopia[i] == basesEvaCopia[i]:
                qubitsBob.append(qubitsEvaCopia[i])
            else:
                if ((qubitsEvaCopia[i] == '|0>') | (qubitsEvaCopia[i] == '|1>')):
                    if (probaQubits <= 0.5):
                        qubitsBob.append('|+>')
                    else:
                        qubitsBob.append('|->')
                elif ((qubitsEvaCopia[i] == '|+>') | (qubitsEvaCopia[i] == '|->')):
                    if (probaQubits <= 0.5):
                        qubitsBob.append('|0>')
                    else:
                        qubitsBob.append('|1>')

        print("Los qubits de Bob son " + str(qubitsBob))

        #Deteccion de Eva
        for i in range(len(bitsListaCopia)):
              if(qubitsBob[i] == qubitsAliceCopia[i]):
                  continue
              else:
                  print("Los qubits que no coinciden entre Alice y Bob después de haber definido los qubits de Eva son:")
                  print("Los qubits en la posición " + str(i+1))
                  print( "El qubit de Alice es "+ str(qubitsAliceCopia[i]) +", el qubit de Eva es " + str(qubitsEvaCopia[i]) + " y el de Bob es " + str(qubitsBob[i]))
                  print("Eva fue detectada en el bit " + str(i+1))
                  mostrarClave = False
                  break

        #Se imprime la clave privada si no se detecta a Eva
        if mostrarClave == True:
            print("Las bases que coinciden con Alice y Bob son " + str(basesAliceCopia))
            print("Los qubits que coinciden con Alice y Bob son " + str(qubitsAliceCopia))
            print("La clave privada de Alice y Bob es " + str(bitsListaCopia))

    #Se limpian las listas
    bitsLista.clear()
    basesAlice.clear()
    qubitsAlice.clear()
    basesEva.clear()
    qubitsEva.clear()
    basesBob.clear()
    qubitsBob.clear()
    bitsListaCopia.clear()
    basesAliceCopia.clear()
    qubitsAliceCopia.clear()
    basesEvaCopia.clear()
    qubitsEvaCopia.clear()
    basesBobCopia.clear()

    regresarAlMenu()
menu()
```
